# Q&A 모듈 테스트(정성평가)
전체 모듈에 대한 성능을 Eye Checking을 통하여 정성평가 합니다.

```python
import sys
import os
import pandas as pd
#한글 폰트 사용
from matplotlib import font_manager,rc
import matplotlib
import matplotlib.pyplot as plt
#폰트 경로
font_path = "C:/Windows/Fonts/H2GTRM.TTF"
#폰트 이름 얻어오기
font_name = font_manager.FontProperties(fname=font_path).get_name()
#font 설정
matplotlib.rc('font',family=font_name)
```

```python
data_path = 'eyecheck_performance'
assert os.path.isdir(data_path), 'EYE Checking 파일이 준비되어 있지 않습니다.'
```

## 정성평가 Eye Checking Dataset
각 Dataset(aihub, aihub+korquad1, korquad1, korquad2) 에 대하여 가장 좋은 성능을 보인 모델들을 정성평가합니다.
따라서 총 4개지 모델에 대한 정성평가를 진행합니다.

```python
print('정성평가 목록')
for item in os.listdir(data_path):
    print('Dataset : {}  \t\t\t  Model : {}'.format(item.split('_')[0], item.split('_')[1]))
```

### 1. korquad1.0 Dataset - hanbert 정성평가

```python
test_df = pd.read_csv(os.path.join(data_path, 'korquad1_hanbert_eyecheck.csv'))
test_df['opinion'] = test_df[['opinion'+str(idx)for idx in range(4) ]].mean(axis=1)
```

```python
total_count = test_df.groupby(['카테고리'])['카테고리'].count()
opinion_has_answer_count = test_df.groupby(['카테고리'])['opinion'].sum()
opinion = (opinion_has_answer_count/total_count)
```

```python
opinion.plot.bar(title="Hanbert(korquad1.0) Eye Cheking 정답 포함률")
```

```python
fig = plt.figure(figsize=(12, 5))

leftover = sum(total_count.values.tolist()) - sum(opinion_has_answer_count.values.tolist())
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = opinion_has_answer_count.index.tolist() + ['누락']
sizes = opinion_has_answer_count.values.tolist() + [leftover]
explode = [0] * len(labels)
explode[-1] = 0.1
ax1 = fig.add_subplot(111)
ax1.set_title('Hanbert(korquad1.0) Eye Cheking 정답 포함률', fontsize=20)
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


plt.show()
```
